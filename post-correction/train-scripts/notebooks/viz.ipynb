```python
import numpy as np

# Replace 'your_file.csv' with the actual path to your CSV file
file_path = '/home/ocr_proj/OCR/post_correction/pe-ocr-sanskrit/train-scripts/notebooks/gcer_results.csv'

# Load the CSV file as a NumPy array
data = np.genfromtxt(file_path, delimiter=',')
```

```python
data.shape
```

```python
import matplotlib.pyplot as plt

# X coordinates
x = [2, 5, 10, 20, 40]

# Colors for each set of points
colors = ['red', 'green', 'blue']

# Create the plot
plt.figure(figsize=(10, 6))

# Plot each set of points
for i in range(3):
    if i == 0:
        label = "CNN-LSTM"
    if i == 1:
        label = "CNN-LSTM + LLM"
    if i == 2:
        label = "Best of Both"
    plt.scatter(x, data[i], color=colors[i], marker='o', linestyle='-', linewidth=2,  label=label)

# Add horizontal line at y = data[0][4]
plt.axhline(y=data[0][4], color='red', linestyle='--', label='pure CNN-LSTM')

# Customize the plot
plt.xlabel('CNN-LSTM Training data size (in Pages). The Remain pages are used to train the LLM', fontsize=12)
plt.ylabel('GCER', fontsize=12)
plt.title('Budgeting 40 Pages of annotated data', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

# Use logarithmic scale for x-axis due to the wide range of x values
plt.xscale('log')
# Set specific x-axis ticks
plt.xticks(x, x)
# Show the plot
plt.tight_layout()
plt.show()
```
